```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

```

```python
df = pd.read_csv(r"C:\Users\Tien\Desktop\KHDL\data-science\data_science\covid.csv")
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'])
#df = df.fillna(0)
```

```python
year_in = int(input("Nhập năm cần xem: "))
nuoc = input("Nhập quốc gia cần xem: ")
```

```python
a = np.zeros(12)
b = np.zeros(12)
c = np.zeros(12)
data_np = df.to_numpy()
for Date,Country,Province,Confirmed,Recovered,Deaths in data_np:
    if Date.year == 2021 and Date.is_month_end:
        a[Date.month - 1] += Confirmed
        b[Date.month - 1] += Deaths
        c[Date.month - 1] += Recovered
print(a)
print(b)
print(c)
```

```python
fig,(x1,x2,x3) = plt.subplots(1,3,figsize = (20,10), dpi=120)
x1.plot(a,'r',label = 'Nhiễm')
x1.legend()
x2.plot(b,'y', label = 'Chết' )
x2.legend()
x3.plot(c,'b', label = 'Hồi phục')
x3.legend()
```
